```python
# Read data from csv with delimiter parameter
dataset = pd.read_csv('bank_telesale_exercise.csv', sep=';') # sep parameter default it ","
# print out the dataset
dataset
```

```python
# print out list of columns of the dataset
dataset.columns
```

```python
# remove out the "Unnamed: 0" column
# we actually assign new copy of the dataset without the columns which have name starting with "Unnamed"
dataset = dataset.loc[:, ~dataset.columns.str.contains('^Unnamed')] 
# print out the dataset
dataset
```

### Let start play around with dataset

```python
x = dataset['age'].unique() # get unique value of age
x.sort() # sort the array in order
print("Quantity of unique value of age variable: %s \n" %(len(x))) # show quantity of value
print('The array of values of age: \n{} \n'.format(x)) # show the array
y_left = dataset.groupby('age')['age'].count() # group by base on age column then count for each value
print('The result counting of values of age: \n{} \n'.format(y_left)) # show the result of group by statement
y_right = dataset.groupby('age')['current_outcome'].apply(lambda x: x[x == 'yes'].count()/x.count()) # group by base on age column then calculate percentage of current_outcome column
print('The result calculate convertion rate base on age: \n{} \n'.format(y_right)) # show the result of group by statement
```

```python
# Plot age column in histograms chart
plt.hist(dataset['age'], bins=64)

plt.show() # if there is plt.show() here the last plt.show() will be used and stack all chart into only one

# Plot age column in bar chart
plt.bar(x, y_left)

plt.show() # then we should call plt.show() for every single chart

# Plot age & convertion rate in two bar chart 
fig, axs = plt.subplots(2,1)

axs[0].bar(x, y_left)
axs[1].bar(x, y_right)

# plt.show() # it is not necessary due to not using plt to draw chart

# Plot age & convertion rate in same chart bar for age & scatter for convertion rate
fig = plt.figure()
ax_left = fig.add_subplot(111) # alternative to fig.add_subplot(1,1,1) => nrows = 1; ncols = 1, index = 1 ! matplotlib 3.1 can use fig.add_subplot() instead
ax_right = ax_left.twinx() # ax_right will share x_axis with ax_left; twiny for share y_axis

ax_left.bar(x, y_left[x], color='blue') # in matplotlib 3.0 use color='b' instead
ax_right.scatter(x, y_right[x], color='red') # in matplotlib 3.0 use color='r' instead

# plt.show() # it is not necessary
```
